```python
#Import Essential Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import dataset
df=pd.read_csv("student_info.csv")
df.head()
```

```python
#Visualize the data
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.title("Student Marks Predictor")
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.show()
```

```python
#Handling Missing Values
#1.Check Missing Values
df.isnull().sum()
#2.Filling Missing Values
```

```python
df=df.fillna(df.mean())
```

```python
#Again Checking Missing Values
df.isnull().sum()
```

```python
df.head()
```

```python
#Separate dataset into independent and dependent variable
#x=Independant variable.
#y=dependant variable.
x=df.drop("student_marks",axis=1)
y=df.drop("study_hours",axis=1)
x.head()
```

```python
#Train-Test Splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
#Apply Linear Regression algo
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Test Model
import warnings
warnings.filterwarnings("ignore")
sh=float(input("Enter The Study Hours<=12: "))
if sh<=12:
         res=lr.predict([[sh]])[0][0].round(2)
         print("Expected marks=",res)
else:
    print("Invalid Input")
```
