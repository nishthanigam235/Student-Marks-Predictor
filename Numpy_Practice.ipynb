```python
import numpy as np
```

```python
a=np.array([1,2,3])
print(a)
```

```python
#2D Array
b=np.array([(1,2,3),(4,5,6)])
print(b)
```

```python
#Use of ndim property of numpy array
print(a.ndim)
print(b.ndim)
```

```python
#To check datatype of numpy array=dtype
print(a.dtype)
print(b.dtype)
```

```python
#To use the item size
print(a.itemsize)
print(b.itemsize)
```

```python
#Shape tells no of rows and columns and size will count number of element
print(a.size)
print(b.size)
print(a.shape)
print(b.shape)
```

```python
#Reshape is used to change dimensions
print(b.reshape(3,2))
```

```python
#Linspace=returns evenly spaced numbers over a specified interval
c=np.linspace(1,3,50)
print(c)
```

```python
#Use of min ,sum and max function
print("Maximum value=",a.max())
print("Minimum value=",a.min())
print("Sum of values=",a.sum())
```
